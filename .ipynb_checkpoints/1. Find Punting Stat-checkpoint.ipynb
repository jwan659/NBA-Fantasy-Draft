Get nba player stats from the past three seasons.

## Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('Data/2021-Player-Data.csv')
df.info()
```

```python
corr_columns = ['FG%','FT%','3P','PTS','TRB','AST','STL','BLK','TOV']
corr_df = df[corr_columns]
corr_df.info()
```

```python
fig, ax = plt.subplots(figsize=(10,10))
sns_plot = sns.heatmap(corr_df.corr(), annot=True, cmap='coolwarm', fmt='.2g', ax=ax)
fig = sns_plot.get_figure()
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

def calc_vif(X):

    # Calculating VIF
    vif = pd.DataFrame()
    vif["variables"] = X.columns
    vif["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]

    return(vif)
```

```python
calc_vif(corr_df)
```
